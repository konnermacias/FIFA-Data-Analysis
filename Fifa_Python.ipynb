```python
train = train.drop(['Ob', 'Real Face', 'Loaned From'], axis = 1)
test = test.drop(['Ob', 'Real Face', 'Loaned From'], axis = 1)
```

```python
train['Preferred Foot'] = train['Preferred Foot'].fillna(train['Preferred Foot'].dropna().mode()[0])
test['Preferred Foot'] = test['Preferred Foot'].fillna(train['Preferred Foot'].dropna().mode()[0])
train['Internation Reputation'] = train['International Reputation'].fillna(np.around(train['International Reputation'].dropna().mean()))
test['Internation Reputation'] = test['International Reputation'].fillna(np.around(train['International Reputation'].dropna().mean()))
train['Weak Foot'] = train['Weak Foot'].fillna(np.around(train['Weak Foot'].dropna().mean()))
test['Weak Foot'] = test['Weak Foot'].fillna(np.around(train['Weak Foot'].dropna().mean()))
train['Skill Moves'] = train['Skill Moves'].fillna(np.around(train['Skill Moves'].dropna().mean()))
test['Skill Moves'] = test['Skill Moves'].fillna(np.around(train['Skill Moves'].dropna().mean()))
train['Work Rate'] = train['Work Rate'].fillna(train['Work Rate'].dropna().mode()[0])
test['Work Rate'] = test['Work Rate'].fillna(train['Work Rate'].dropna().mode()[0])
train['Body Type'] = train['Body Type'].fillna(train['Body Type'].dropna().mode()[0])
test['Body Type'] = test['Body Type'].fillna(train['Body Type'].dropna().mode()[0])
train['Position'] = train['Position'].fillna(train['Position'].dropna().mode()[0])
test['Position'] = test['Position'].fillna(train['Position'].dropna().mode()[0])
#Issue with International Reputation
```
